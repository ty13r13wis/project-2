```python

# Finding #1: Most students heard about computing classes through word of mouth
fig1, ax1 = plt.subplots()
df['how did you hear about county college of morris? [family member or friend]'].value_counts().plot(kind='bar', ax=ax1)
ax1.set_title("How Did Students Hear About Computing Classes?")
ax1.set_xlabel("Family or Friend")
ax1.set_ylabel("Number of Students")
```

```python


#Finding #2 a high amount of students from the shareTime program are interested in computing classes

# Filter out rows where the 'degree' column is IT or CS
df_filtered = df.loc[~df['what degree program are you currently enrolled in?'].isin(['Information Technology', 'Computer Science'])]

# Count the number of students in each major
major_counts = df_filtered['what degree program are you currently enrolled in?'].value_counts()

# Create a bar chart
plt.bar(major_counts.index, major_counts.values)
plt.xlabel('Non Computing major students')
plt.ylabel('Number of Students')
plt.title('Number of Students per Degree Program (excluding IT and CS)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Finding #3 what was the gender of most of the students after cleaning the data##
# Distribution of gender
gender_count = df['gender'].value_counts()
plt.bar(gender_count.index, gender_count.values)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Finding #4 Age distribution after cleaning the data##
#plt.title('Age Distribution')
#plt.xlabel('age')
#plt.ylabel('Count')
#plt.show()

plt.figure(figsize=(10, 6)) # Set the figure size to be wider
plt.hist(df['age'])
plt.title('Age Distribution')
plt.xlabel('age')
plt.xticks(range(0, max(df('age')+5, 5)) # Set the x-ticks to be at intervals of 5
plt.show()
```
